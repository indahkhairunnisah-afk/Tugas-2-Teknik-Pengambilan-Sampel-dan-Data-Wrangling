#Persiapan Dan Load Dataset

Langkah awal adalah memuat library yang dibutuhkan dan membaca dataset. Kita juga perlu mengubah karakter ? menjadi NaN agar library Pandas dapat mendeteksinya sebagai data yang hilang (missing value).


```python
import pandas as pd
import numpy as np
```

```python
# Memuat dataset
df = pd.read_csv('/content/data_tugas 2.csv')
```

mengubah karakter "?" menjadi NaN agar library Pandas dapat mendeteksinya sebagai data yang hilang (missing value).

```python
# Mengganti karakter '?' dengan NaN agar terdeteksi sebagai missing value
df = df.replace('?', np.nan)
```

```python
# Menampilkan 5 data teratas
df.head()
```

```python
#Menampilkan 5 data terbawah
df.tail()
```

#Menampilkan Jumlah Data Hilang

 Menghitung jumlah data yang kosong pada setiap kolom untuk mengetahui sebaran missing value.

 *Menghasilkan DataFrame boolean (True jika data kosong/NaN, False jika ada nilai).
 Menjumlahkan nilai True per kolom, sehingga diperoleh jumlah data hilang (missing value) di setiap variabel.*

```python
# Menghitung jumlah missing value per variabel
missing_values = df.isnull().sum()
```

*Menampilkan tabel jumlah missing value untuk setiap variabel, sehingga dapat menampilkan distribusi data hilang di seluruh kolom.*

```python
# Membuat tabel ringkasan
print("Tabel Jumlah Missing Value:")
print(missing_values)
```

*Mencari nama kolom dengan nilai maksimum (kolom yang paling banyak missing value). max() mengambil jumlah maksimum missing value tersebut.
Menampilkan Variabel mana yang paling banyak kehilangan data (missing Value), dan berapa jumlahnya.* *teks yang dimiringkan*

```python
# Mencari variabel dengan missing value terbanyak
max_missing_col = missing_values.idxmax()
max_missing_val = missing_values.max()

print(f"\nInterpretasi: Variabel yang memiliki data hilang paling banyak adalah '{max_missing_col}' dengan jumlah {max_missing_val} data.")
```

*Berdasarkan hasil pengecekan, terdapat beberapa variabel yang memiliki data hilang. Variabel normalized-losses memiliki jumlah data hilang paling banyak, yaitu sebanyak 41 data.*

#Mengganti Missing Value Pada Variabel Kategori

*Untuk variabel kategorik (make, bahan-bakar, jumlah-pintu, body-style), data yang hilang akan diganti dengan label baru yaitu "unknown".*

*Memilih variabel 'make', 'bahan-bakar', 'jumlah-pintu', 'body-style' dan mengganti nilai kosong (NaN) dengan unknown.*

```python
# Daftar variabel kategori
cat_cols = ['make', 'bahan-bakar', 'jumlah-pintu', 'body-style']

# Mengganti NaN dengan "unknown"
for col in cat_cols:
    df[col] = df[col].fillna('unknown')
```

*Untuk setiap variabel kategori, ditampilkan jumlah kemunculan tiap kategori, dengan ini terlihat distribusi per kategori termasuk berapa banyak yang masuk kategori 'unknwon'*

```python
print("Ringkasan Jumlah Kategori:")
for col in cat_cols:
    print(f"\nVariabel: {col}")
    print(df[col].value_counts())
```

*Setelah proses pembersihan dilakukan, seluruh variabel kategori sudah bebas dari nilai kosong karena setiap NaN diganti dengan label "unknown". Misalnya, pada variabel bahan-bakar terdapat 13 data yang kini berlabel "unknown", dan pada jumlah-pintu ada 18 data dengan label serupa. Pendekatan ini dipilih agar semua baris data tetap dipertahankan tanpa harus melakukan imputasi yang bersifat subjektif, sehingga analisis tetap konsisten sekaligus transparan terhadap keberadaan data yang tidak lengkap.*

#Mengganti Missing Value pada Variabel Numerik

Untuk variabel numerik (length, width, height), kita akan melakukan imputasi menggunakan nilai rata-rata (mean).

```python
# Daftar variabel numerik
num_cols = ['length', 'width', 'height']

# Mengubah tipe data ke float (untuk memastikan perhitungan mean akurat)
for col in num_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Hitung rata-rata dan isi missing value
means = df[num_cols].mean()
df[num_cols] = df[num_cols].fillna(means)

print("Nilai rata-rata yang digunakan untuk imputasi:")
print(means)
```

*Nilai yang hilang pada data numerik telah diganti dengan rata-rata masing-masing kolom, sehingga variabel length diisi dengan 173.90, width dengan 65.93, dan height dengan 53.73. Pengisian menggunakan rata-rata ini bertujuan untuk menjaga konsistensi distribusi data, sehingga meskipun ada data yang sebelumnya kosong, pola keseluruhan tetap terjaga tanpa menimbulkan bias yang berarti.*

#Min-Max Normalization

 *Melakukan normalisasi pada variabel wheel-bases agar nilainya berada dalam rentang 0 hingga 1.*

*Setiap nilai x pada kolom wheel-bases dikurangi dengan nilai minimum kolom tersebut, Hasilnya kemudian dibagi dengan selisih antara nilai maksimum dan minimum, dengan ini semua nilai akan dipetakan ke rentang 0 sampai 1*

*Hasil normalisasi disimpan dalam kolom baru bernama wheel-bases-normalized, Kolom asli (wheel-bases) tetap ada, sehingga bisa dibandingkan dengan hasil normalisasi.*

```python
# Rumus Min-Max: (x - min) / (max - min)
df['wheel-bases-normalized'] = (df['wheel-bases'] - df['wheel-bases'].min()) / (df['wheel-bases'].max() - df['wheel-bases'].min())
```

*Menampilkan 5 baris pertama dari kolom asli dan kolom hasil normalisasi, tujuannya agar menampilkan bisa melihat bagaimana nilai asli dipetakan ke skala 0-1.*

```python
# Menampilkan perbandingan kolom asli dan hasil normalisasi
print(df[['wheel-bases', 'wheel-bases-normalized']].head())
```

Normalisasi Min-Max berhasil dilakukan pada kolom wheel-bases. Contohnya, nilai asli 88.6 kini telah bertransformasi menjadi sekitar 0.058 dalam skala 0-1. Normalisasi ini sangat berguna untuk algoritma machine learning agar fitur dengan skala besar tidak mendominasi fitur dengan skala kecil.

#Z-Score Normalization

*Melakukan standarisasi (Z-score normalization) pada variabel length, width, dan height sehingga memiliki rata-rata 0 dan standar deviasi 1.*

*Setiap nilai x dikurangi dengan rata-rata (mean) kolom, Hasilnya dibagi dengan standar deviasi (std) kolom tersebut,Nilai Z-score menunjukkan seberapa jauh suatu data dari rata-rata dalam satuan standar deviasi.*

*Dilakukan perulangan untuk tiga variabel: length, width, dan height, Hasil standarisasi disimpan dalam kolom baru dengan nama tambahan -zscore (misalnya length-zscore).*

```python
# Rumus Z-score: (x - mean) / std
for col in ['length', 'width', 'height']:
    df[f'{col}-zscore'] = (df[col] - df[col].mean()) / df[col].std()
```

*Menampilkan 5 baris pertama dari nilai asli dan nilai hasil Z-score untuk dibandingkan.*

```python
# Menampilkan hasil standarisasi
print(df[['length', 'length-zscore', 'width', 'width-zscore', 'height', 'height-zscore']].head())
```

*Dengan standarisasi Z-score, data length, width, dan height kini berada dalam skala yang sama, yaitu rata-rata bernilai 0 dan standar deviasi bernilai 1. Jika nilainya negatif (seperti length-zscore -0.42), berarti nilai tersebut berada di bawah rata-rata populasi, sedangkan nilai positif berarti di atas rata-rata. Hal ini memudahkan analisis statistik maupun pemodelan machine learning, karena variabel dengan skala berbeda tidak lagi mendominasi perhitungan.*
